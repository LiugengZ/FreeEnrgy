```python

class PingPongEnv:
    def __init__(self):
        self.state = np.zeros((10, 10))  # 初始化一个10x10的网格
        self.ball_position = (1, 5)  # 球的初始位置
        self.moveRow = -1
        self.moveCol = -1
        self.paddle_row = 5  # 拍子位置
        self.paddle_size = 3  # 拍子大小
        self.movepaddle = 0

    def step(self,move):

        #new_row = self.
        
        # 更新球的位置
        if (self.ball_position[0] !=0 and self.ball_position[1]!=0 and self.ball_position[0] !=9 and self.ball_position[1]!=9):
            self.moveCol = self.moveCol;
            self.moveRow = self.moveRow;
        elif (self.ball_position[0] ==0  and (self.ball_position[1]!=0 or self.ball_position[1]!=9)):
            self.moveRow = -self.moveRow;
        elif (self.ball_position[0] ==9):
            if  self.ball_position[1]<(self.paddle_row + (self.paddle_size/2)) and self.ball_position[1]> (self.paddle_row - (self.paddle_size/2)):
                self.moveRow = -self.moveRow
            else:
                return self._get_observation(True),0 
        elif ((self.ball_position[0] !=0  or self.ball_position[0] !=9) and (self.ball_position[1]==0 or self.ball_position[1]==9)):
            self.moveCol = -self.moveCol;
        else:
            self.moveCol = -self.moveCol;
            self.moveRow = -self.moveRow;
            
        self.ball_position =  (self.ball_position[0] + self.moveRow, #np.random.choice([-1, 0, 1]),
                              self.ball_position[1] + self.moveCol)#np.random.choice([-1, 0, 1]))

        # 确保球在网格内
        self.ball_position = (int(max(0, min(9, self.ball_position[0]))),
                              int(max(0, min(9, self.ball_position[1]))))
        self.state[self.ball_position] = 1  # 标记球的位置
        self.movepaddle = move #拍子位置
        self.paddle_row = self.paddle_row + self.movepaddle #拍子位置
        if self.paddle_row < 0 or self.paddle_row>9:
            return self._get_observation(True),0
        #print(self.paddle_row)
        return self._get_observation(True),1

        

    def reset(self):
        ball_y = np.random.randint(1, 8)
        self.state = np.zeros((10, 10))
        self.ball_position = (1, ball_y)
        self.paddle_row = 5
        self.moveRow = -1
        self.moveCol = -1
        self.movepaddle = 0
        return self._get_observation(True)

    def _get_observation(self, caught):
        if caught:
            observation = [self.ball_position[1], self.paddle_row]
        else:
            random_value = np.random.randint(0, 10)
            observation = [random_value, self.paddle_row]
        return observation

    def checkEnd(self,n):
        if  not n :
            self.reset()
```

```python
class Cell:
    def __init__(self, id):
        self.id = id
        self.state = 0.0
        self.connections = {}  # key: cell id, value: weight
        self.output = 0.0

    def activate(self):
        self.output = max(0, self.state)

    def add_connection(self, cell_id, weight):
        self.connections[cell_id] = weight

    def prune_connections(self, threshold):
        to_remove = [cell_id for cell_id, weight in list(self.connections.items()) if abs(weight) < threshold]
        for cell_id in to_remove:
            del self.connections[cell_id]
```

```python
class Network:
    def __init__(self):
        self.cells = {}  # key: cell id, value: cell object
        self.connections = {}  # key: (cell_id_from, cell_id_to), value: weight

    def add_cell(self, cell):
        self.cells[cell.id] = cell

    def connect(self, cell_id_from, cell_id_to, weight):
        if cell_id_from in self.cells and cell_id_to in self.cells:
            self.cells[cell_id_from].add_connection(cell_id_to, weight)
            self.connections[(cell_id_from, cell_id_to)] = weight

    def send_signals(self):
        signals = {}
        for cell in self.cells.values():
            for connected_id, weight in cell.connections.items():
                if connected_id not in signals:
                    signals[connected_id] = 0.0
                signals[connected_id] += cell.output * weight
        return signals

    def update_all(self, external_inputs):
        # 应用外部输入
        for cell_id, input_signal in external_inputs.items():
            if cell_id in self.cells:
                self.cells[cell_id].state += input_signal

        # 传递内部信号
        internal_signals = self.send_signals()
        for cell_id, signal in internal_signals.items():
            if cell_id in self.cells:
                self.cells[cell_id].state += signal

        # 更新每个单元的状态
        for cell in self.cells.values():
            cell.activate()
```

```python
class AdaptiveNetwork(Network):
    def __init__(self, growth_threshold=0.5):
        super().__init__()
        self.growth_threshold = growth_threshold

    def add_cell(self, cell):
        super().add_cell(cell)

    def grow(self, cell_id):
        if self.cells[cell_id].output > self.growth_threshold:
            new_cell = Cell(len(self.cells))
            self.add_cell(new_cell)
            self.connect(cell_id, new_cell.id, 0.1)
            self.connect(new_cell.id, cell_id, 0.1)

    def update_all_with_growth(self, external_inputs):
        self.update_all(external_inputs)
        # Collect cells to grow
        cells_to_grow = []
        for cell_id, cell in list(self.cells.items()):
            if cell.output > self.growth_threshold:
                cells_to_grow.append(cell_id)
        # Apply growth after iteration
        for cell_id in cells_to_grow:
            self.grow(cell_id)
        # Prune connections
        cells_to_prune = list(self.cells.values())
        for cell in cells_to_prune:
            cell.prune_connections(threshold=0.05)
```

```python
def train_network(network, environment, episodes=1000):
    for episode in range(episodes):
        # 重置环境
        observation = environment.reset()
        # 生成外部输入，根据是否接住球决定输入信号
        if episode % 2 == 0:
            external_inputs = {cell_id: observation[0] for cell_id in network.cells.keys()}
        else:
            external_inputs = {cell_id: np.random.rand() for cell_id in network.cells.keys()}
        # 更新网络
        network.update_all_with_growth(external_inputs)
        # 获取网络输出，控制球拍
        control_signal = np.mean([cell.output for cell in list(network.cells.values())])
        action = control_signal - 0.5
        # 执行动作，获取奖励
        observation, reward = environment.step(action)
        # 根据奖励调整连接权重
        for cell in list(network.cells.values()):
            for connected_id, weight in list(cell.connections.items()):
                post_cell = network.cells.get(connected_id)
                if post_cell:
                    # Hebbian learning rule
                    delta_w = 0.01 * reward * cell.output * post_cell.output
                    cell.connections[connected_id] += delta_w
        print(f"Episode {episode}, Reward: {reward}, Cells: {len(network.cells)}, Connections: {sum(len(cell.connections) for cell in network.cells.values())}")
```

```python
def train_network(network, environment, episodes=1000):
    for episode in range(episodes):
        # 重置环境
        environment.reset()
        # 生成外部输入，根据是否接住球决定输入信号
        observation, _ = environment.step(0)  # 获取初始观察值
        if episode % 2 == 0:
            external_inputs = {cell_id: observation[0] for cell_id in network.cells.keys()}
        else:
            external_inputs = {cell_id: np.random.rand() for cell_id in network.cells.keys()}
        # 更新网络
        network.update_all_with_growth(external_inputs)
        # 获取网络输出，控制球拍
        control_signal = np.mean([cell.output for cell in network.cells.values()])
        action = control_signal - 0.5
        # 执行动作，获取奖励
        observation, reward = environment.step(action)
        # 根据奖励调整连接权重
        for cell in list(network.cells.values()):
            for connected_id, weight in list(cell.connections.items()):
                post_cell = network.cells.get(connected_id)
                if post_cell:
                    # Hebbian学习规则
                    delta_w = 0.01 * reward * cell.output * post_cell.output
                    cell.connections[connected_id] += delta_w
        print(f"Episode {episode}, Reward: {reward}, Cells: {len(network.cells)}, Connections: {sum(len(cell.connections) for cell in network.cells.values())}")
```
